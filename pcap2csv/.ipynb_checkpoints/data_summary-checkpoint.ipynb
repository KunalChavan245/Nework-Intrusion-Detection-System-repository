```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.DataFrame({'a':list(range(100000)),
              'b':list(range(100000)),
              'c':list(range(100000))})
```

```python
df = pd.read_csv('/home/neto/Desktop/IoT_attack_data/merged_files/DDoS-SlowLoris.pcap.csv')
```

```python
df_summary_list = []
```

```python
n_rows = 10
```

```python
last_row = 0
while last_row<len(df):
    if last_row%10000==0:
        print('row = ', last_row)
    sliced_df = df[last_row:last_row+n_rows]
    sliced_df = pd.DataFrame(sliced_df.mean()).T# mean
    df_summary_list.append(sliced_df)
    last_row += n_rows
```

```python
df_summary = pd.concat(df_summary_list).reset_index(drop=True)
```

```python
df_summary.to_csv('/home/neto/Desktop/IoT_attack_data/merged_files/Uploading_Attack.pcap_summary10.csv')
```
